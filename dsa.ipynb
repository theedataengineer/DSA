## Big O Notation

- Big O Notation - is used to measure how running time or space requirements for your program grow as input size grows.

Note: As the size of the array grows, the time also grows.
- Normally, it's a linear function.
- time =  a*n + b
- So, as a grows and b remains constant: time = a*n
- We then drop the constants: time = 0(n)

- That mathematical representation is the Big O Notation

### BigO
- BigO refers to very large value of n. Hence, if you have a function like:

time = 5*n^2 + 3*n + 20

- When value of n is very large b*n +c become irrelevant.

example: n = 1000

- Time Complexity

## Linked List

- Remember, Arrays store values in contiguous memory location
- Linked Lists store values at random memory locations, but they are linked by pointers.

### NOTE:

- Insert Element at beginning = 0(1)
- Delete Element at beginning = 0(1)
- Insert/Delete at beginning = 0(n)


#### Linked List Traversal = 0(n)
#### Accessing Element By Value = 0(n)

### Benefits of Linked List over an Array

1. You don't need to pre-allocate space
2. Insertion is easier
